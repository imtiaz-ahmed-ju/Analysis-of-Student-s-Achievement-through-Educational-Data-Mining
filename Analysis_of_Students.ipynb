```python
df["SPL"].replace("A+", "4.00", inplace=True)
df["SPL"].replace("A", "3.75", inplace=True)
df["SPL"].replace("A-", "3.50", inplace=True)
df["SPL"].replace("B+", "3.25", inplace=True)
df["SPL"].replace("B", "3.00", inplace=True)
df["SPL"].replace("B-", "2.75", inplace=True)
df["SPL"].replace("C+", "2.50", inplace=True)
df["SPL"].replace("C", "2.25", inplace=True)
df["SPL"].replace("D", "2.00", inplace=True)
df["SPL"].replace("0", "0.00", inplace=True)
```

```python
df['SPL'] = df['SPL'].replace(['A+','A','A-', 'B+','B','B-','C+', 'C','D'],[4.00,3.75,3.50, 3.25,3.00,2.75,2.50, 2.25,2.00])
```

```python
df['EEC'] = df['EEC'].replace(['A+','A','A-', 'B+','B','B-','C+', 'C','D'],[4.00,3.75,3.50, 3.25,3.00,2.75,2.50, 2.25,2.00])
```

```python
df['Calculus'] = df['Calculus'].replace(['A+','A','A-', 'B+','B','B-','C+', 'C','D'],[4.00,3.75,3.50, 3.25,3.00,2.75,2.50, 2.25,2.00])
```

```python
df['Physics'] = df['Physics'].replace(['A+','A','A-', 'B+','B','B-','C+', 'C','D'],[4.00,3.75,3.50, 3.25,3.00,2.75,2.50, 2.25,2.00])
```

```python
df['English'] = df['English'].replace(['A+','A','A-', 'B+','B','B-','C+', 'C','D'],[4.00,3.75,3.50, 3.25,3.00,2.75,2.50, 2.25,2.00])
```

```python
df
```

```python
df.dtypes
```

```python
df['CGPA']= df['CGPA']
conditions = [
    (df['CGPA'] >= 3.50) & (df['CGPA']<= 4.00),
    (df['CGPA'] >= 3.00) & (df['CGPA'] < 3.5 ),
    (df['CGPA'] >= 2.5) & (df['CGPA'] < 3.00 ),
    (df['CGPA'] >= 2.00) & (df['CGPA'] < 2.5 ),
    (df['CGPA'] < 2.00)
    ]
choices = ['1', '2', '3', '4', '5']
df['CGPA'] = np.select(conditions, choices, default='0')
print(df)
```
